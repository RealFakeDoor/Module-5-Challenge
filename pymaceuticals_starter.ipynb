```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "./data/Mouse_metadata.csv"
study_results_path = "./data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
complete_mouse_data = pd.merge(study_results,mouse_metadata, on = 'Mouse ID', how = 'left')

# Display the data table for preview
complete_mouse_data
```

```python
# Checking the number of mice.
mouse_count = complete_mouse_data['Mouse ID'].nunique()
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse_ID_g989 = complete_mouse_data[complete_mouse_data['Mouse ID'] == 'g989']
mouse_ID_g989['Mouse ID'].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouse_ID_g989
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
    # Do we just drop the duplicates or do we drop the entire dataset?
# clean_mouse_data = complete_mouse_data.drop('g989', axis = 1)
clean_mouse_data = complete_mouse_data[complete_mouse_data['Mouse ID'] != 'g989']
clean_mouse_data
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count = clean_mouse_data['Mouse ID'].nunique()
mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen.
Drug_regimen_complete = clean_mouse_data.groupby('Drug Regimen')

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
    # mean, median, variance, standard deviation, and SEM of the tumor volume. 
    # Assemble the resulting series into a single summary DataFrame.
Summary_Table = pd.DataFrame({'Mean Tumor Volume': Drug_regimen_complete['Tumor Volume (mm3)'].mean(),
                             'Median Tumor Volume': Drug_regimen_complete['Tumor Volume (mm3)'].median(),
                             'Tumor Volume Variance' : Drug_regimen_complete['Tumor Volume (mm3)'].var(),
                             'Tumor Volume Std. Dev.': Drug_regimen_complete['Tumor Volume (mm3)'].std(),
                             'Tumor Volume Std. Err.': Drug_regimen_complete['Tumor Volume (mm3)'].agg(st.sem)
                             })
Summary_Table
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
Summary_Table = pd.DataFrame(Drug_regimen_complete.agg(
                                          mean=('Tumor Volume (mm3)', 'mean'), 
                                          median=('Tumor Volume (mm3)', 'median'),
                                          variance = ('Tumor Volume (mm3)','var'),
                                          std = ('Tumor Volume (mm3)','std'),
                                          sem = ('Tumor Volume (mm3)','sem')
                                         ))
Summary_Table
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
y_values = Drug_regimen_complete['Timepoint'].count()

y_values.sort_values(ascending=False, inplace=True)

# Create a bar plot & Set the title and labels:
y_values.plot(kind='bar', figsize=(10, 6), 
                  title = 'Rows (Mouse ID/Timepoints) within Each Drug Regimen Group',
                 xlabel = 'Drug Regimen',
                 ylabel = '# of Observed Mouse Timepoints')
# Show the plot
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
plt.bar(y_values.index, y_values.values)
plt.xticks(rotation=90, ha='right')

plt.title('Rows (Mouse ID/Timepoints) within Each Drug Regimen Group')
plt.xlabel('Drug Regimen')
plt.ylabel('# of Observed Mouse Timepoints')
```
