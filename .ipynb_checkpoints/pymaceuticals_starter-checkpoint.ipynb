```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "./data/Mouse_metadata.csv"
study_results_path = "./data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
complete_mouse_data = pd.merge(study_results,mouse_metadata, on = 'Mouse ID', how = 'left')

# Display the data table for preview
complete_mouse_data
```

```python
# Checking the number of mice.
mouse_count = complete_mouse_data['Mouse ID'].nunique()
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse_ID_g989 = complete_mouse_data[complete_mouse_data['Mouse ID'] == 'g989']
mouse_ID_g989['Mouse ID'].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouse_ID_g989
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
    # Do we just drop the duplicates or do we drop the entire dataset?
# clean_mouse_data = complete_mouse_data.drop('g989', axis = 1)
clean_mouse_data = complete_mouse_data[complete_mouse_data['Mouse ID'] != 'g989']
clean_mouse_data
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count = clean_mouse_data['Mouse ID'].nunique()
mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen.
Drug_regimen_complete = clean_mouse_data.groupby('Drug Regimen')

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
    # mean, median, variance, standard deviation, and SEM of the tumor volume. 
    # Assemble the resulting series into a single summary DataFrame.
Summary_Table = pd.DataFrame({'Mean Tumor Volume': Drug_regimen_complete['Tumor Volume (mm3)'].mean(),
                             'Median Tumor Volume': Drug_regimen_complete['Tumor Volume (mm3)'].median(),
                             'Tumor Volume Variance' : Drug_regimen_complete['Tumor Volume (mm3)'].var(),
                             'Tumor Volume Std. Dev.': Drug_regimen_complete['Tumor Volume (mm3)'].std(),
                             'Tumor Volume Std. Err.': Drug_regimen_complete['Tumor Volume (mm3)'].agg(st.sem)
                             })
Summary_Table
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
Summary_Table = pd.DataFrame(Drug_regimen_complete.agg(
                                          mean=('Tumor Volume (mm3)', 'mean'), 
                                          median=('Tumor Volume (mm3)', 'median'),
                                          variance = ('Tumor Volume (mm3)','var'),
                                          std = ('Tumor Volume (mm3)','std'),
                                          sem = ('Tumor Volume (mm3)','sem')
                                         ))
Summary_Table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin




# Start by getting the last (greatest) timepoint for each mouse
Timepoint_ID_mouse_data =  complete_mouse_data.loc[:, ['Mouse ID', 'Timepoint']]

# Timepoint_ID_mouse_data =  clean_mouse_data.loc[:, ['Mouse ID', 'Timepoint']]
Mice_ID = Timepoint_ID_mouse_data.groupby('Mouse ID').max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
df = pd.merge(Mice_ID, clean_mouse_data, on= ['Mouse ID', 'Timepoint'], how='inner')
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']


    
# Create empty list to fill with tumor vol data (for plotting)
tumor_vols = []

# Calculate the IQR and quantitatively determine if there are any potential outliers:
for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes:
        # Append each series into the Tumor Volume list.
    subset = df.loc[df['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']
    
    # add subset 
    tumor_vols.append(subset)
    
    # Determine outliers using upper and lower bounds.
    Q1, Q2, Q3 = subset.quantile([0.25, 0.5, 0.75])
    
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = subset[(subset < lower_bound) | (subset > upper_bound)]
    print(f"{treatment}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(tumor_vols, 
            labels = treatments, 
            flierprops= dict(markerfacecolor='red', marker='o', markersize= 13))

plt.ylabel('Final Tumor Volume (mm3)')

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin.
Mouse_ID_complete_mouse_data = complete_mouse_data.set_index('Mouse ID')

Capomulin_I509_mouse_data = Mouse_ID_complete_mouse_data.loc['l509', ['Tumor Volume (mm3)', 'Timepoint']]

plt.plot( Capomulin_I509_mouse_data['Timepoint'],Capomulin_I509_mouse_data['Tumor Volume (mm3)'])

plt.ylabel('Tumor Volume (mm3)')
plt.xlabel('Timepoint (days)')
plt.show()
```

```python
# Generate a scatter plot of 
# mouse weight vs. the average observed tumor volume for the entire Capomulin regimen.

# Need to filter values by the drug regimen:
    # set index
Drug_Regimen_complete_mouse_data = complete_mouse_data.set_index('Drug Regimen')
    # Take rows associated with each drug regimen,
    # Columns for Tumor Volume, Weight and Mouse ID.
Capomulin_Tumor_Weight_mouse_data = Drug_Regimen_complete_mouse_data.loc['Capomulin', 
                                    ['Tumor Volume (mm3)', 'Weight (g)', 'Mouse ID']]
    # Need to group the mouse by their ID to ensure we are getting the means by mouse.
group_Capomulin_Tumor_Weight_mouse_data = Capomulin_Tumor_Weight_mouse_data.groupby('Mouse ID')

    # Plot the mean Tumor Volume for each mouse vs the mean weight:
        # find the mean tumor volume for each mouse
mean_tumor = group_Capomulin_Tumor_Weight_mouse_data['Weight (g)'].mean()
        # Need x and Y dimensions to be the same.
        # Assuming the weight stays the same for each mouse, or fluctuates slightly.
        # find mean weight of each mouse.
mean_weight = group_Capomulin_Tumor_Weight_mouse_data['Tumor Volume (mm3)'].mean()

# Plot on scatter
plt.scatter(mean_tumor,  mean_weight)

# Format
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(mean_tumor,mean_weight)

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(mean_tumor, mean_weight)

regress_values = slope * mean_tumor + intercept

line_eq = "y = "+str(round(slope,2))+"*x+"+str(round(intercept,2))


print(f'The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}')

plt.scatter(mean_tumor,  mean_weight)
plt.plot(mean_tumor, regress_values, 'r-')
plt.annotate(line_eq, (5.8, 0.8), color = 'red')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

plt.show()
```
